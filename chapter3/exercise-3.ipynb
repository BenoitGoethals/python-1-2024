Write a function called triangle that takes a string and an integer and draws a pyramid with the given height, made up using copies of the string. Here’s an example of a pyramid with 5 levels, using the string 'L'.


```python
def triangle(letter,count):
    result = ""
    for i in range(1,count+1):
        result = result + (letter *i) + "\n"
    return result

def triangle_2(s:str, height):  
    total = ""
    for i in range(1, height + 1):      
        spaces = " " * (height - i)      
        chars = s * (2 * i - 1)       
        total= total + spaces + chars + "\n"
    return total

def triangle_3(s: str, height: int):  
    return "\n".join(
        " " * (height - i) + s * (2 * i - 1) for i in range(1, height + 1)
    ) + "\n"


print(triangle_2("L", 6))

print(triangle("L",5))
```

Write a function called rectangle that takes a string and two integers and draws a rectangle with the given width and height, made up using copies of the string. Here’s an example of a rectangle with width 5 and height 4, made up of the string 'H'.

rectangle('H', 5, 4)



```python
def rectangle(letter,height,width):
    result = ""
    for i in range(1,height+1):
        result = result + letter * width +"\n"
    return result   

print(rectangle("H",5,4))
```

Write a function named right_justify that takes a string named s as a parameter and prints the string with enough leading spaces so that the last letter of the string is in column 70 of the display.

```python
def right_justify(name):
    return " " * (70-(len(name))) + name

print(right_justify("love "))
print(len(right_justify("love ")))

assert len(right_justify("love ")) == 70
```
